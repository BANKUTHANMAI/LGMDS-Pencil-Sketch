```python
import cv2
from PIL import Image
from IPython.display import display
```

```python
img = cv2.imread('/content/Taj_Mahal.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(img))
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_img))
```

```python
invert_img = cv2.bitwise_not(gray_img)
display(Image.fromarray(invert_img))
```

```python
blur_img = cv2.GaussianBlur(invert_img, (21,21), 0)
display(Image.fromarray(blur_img))
```

```python
inverted_blurred=cv2.bitwise_not(blurred_img)
pencil_sketch = cv2.divide(gray_img, invert_img, scale=250.0)
display(Image.fromarray(pencil_sketch))
```
